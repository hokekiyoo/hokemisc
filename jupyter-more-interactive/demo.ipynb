## デモ
記事のおまけ。

- ブックマークN以下の記事を抽出
- そこからブックマークするユーザーと考え合わせて、記事類似度を抽出
- 記事類似度から記事を2Dマップに落とし込み、可視化

```python
from related_articles import *
import matplotlib.pyplot as plt
from ipywidgets import interact
# 元のURLを表示
root_url = "http://www.procrasist.com"
# url抽出
urls = extract_urls(root_url)
# 記事タイトル抽出
N = 1
titles = []
filtered_urls = []
for url in urls:
    try:
        html = request.urlopen(url)
        data = request.urlopen("http://b.hatena.ne.jp/entry/json/{}".format(url)).read().decode("utf-8")
#         print(bytes(data).strip(b'(').rstrip(b')'))
    except urllib.error.HTTPError as e: 
        # HTTPレスポンスのステータスコードが404, 403, 401などの例外処理
        print(e.reson)
    except urllib.error.URLError as e: 
        # アクセスしようとしたurlが無効なときの例外処理
        print(e.reson)
    #bookmark取得
    try:
        info = json.loads(data.strip('(').rstrip(')'), "r")
    except:
        info = json.loads(data.strip('(').rstrip(')'))
    try:
        num_bookmark = info["count"]
    except:
        continue
    if num_bookmark < N and num_bookmark>3:
        continue
    soup = BeautifulSoup(html, "html.parser")
    title = soup.find("title").text
    print(num_bookmark, title)
    titles.append(title)
    filtered_urls.append(url)

# userリスト作成
print(len(filtered_urls))
M = make_matrix(filtered_urls, save=False)


#記事タイトル抽出
def plot(M, alpha):
    confidences = np.zeros((len(M),len(M)))
    for i, article0 in enumerate(M):
        for j, article1 in enumerate(M):
            a0a1 = np.zeros(len(article0))
            for l,(u0, u1) in enumerate(zip(article0, article1)):
                a0a1[l] = u0*u1
            if article0.sum() == 0 or i==j:
                confidences[i][j] = 0
            else:
                confidences[i][j] = a0a1.sum()#/article0.sum() 
    # symmetricな計量の場合
    dist = 1-np.power(confidences,alpha)
    # 変換
    mds  = manifold.MDS(n_components=2, dissimilarity="precomputed")
    pos = mds.fit_transform(dist)
    return pos
```

```python
from IPython.display import display
import ipywidgets
from bokeh.io import output_notebook, push_notebook
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.models import HoverTool,PanTool,WheelZoomTool,BoxZoomTool,ResetTool,UndoTool,RedoTool

# d = ipywidgets.FloatSlider(value=1, min=0, max=1.0, step=0.01)
d = ipywidgets.BoundedFloatText(value=0.2, min=0, max=1.0, step=0.001,)
pos0 = plot(M, 1.0)
source = ColumnDataSource(data=dict(
    x=pos0[:,0],
    y=pos0[:,1],
    title=titles
))
hover = HoverTool(tooltips=[
    ("index:", "$index"),
    ("(x,y):", "($x,$y)"),
    ("title:", "@title")
])
pan = PanTool()
wheel = WheelZoomTool()
box = BoxZoomTool()
undo = UndoTool()
redo = RedoTool()
reset = ResetTool()

def on_value_change(change):
    alpha = change["new"]
    pos = plot(M, alpha)
    x = pos[:,0]
    y = pos[:,1]
    source.data = dict(x=x,y=y,title=titles)
    push_notebook(handle=t)    

d.observe(on_value_change, names="value")

fig = figure(width=600, height=600, tools=[hover,box,pan,wheel,undo,redo,reset], title="demo")
r = fig.circle("x", "y", source=source, size = 15)
output_notebook()
display(d)
t = show(fig, notebook_handle=True)
```
